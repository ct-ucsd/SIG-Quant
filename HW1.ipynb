```python
import numpy as np
from math import pi
from math import tan
```

```python
def bisection(f, a, b, tol):
    # bisection method of estimating roots
    if np.sign(f(a)) == np.sign(f(b)):
        return print("The scalars a and b do not bound a root")
    m = (a+b)/2
    if np.abs(f(m)) < tol:
          return m
    elif np.sign(f(a)) == np.sign(f(m)):
          return bisection(f, m, b, tol)
    elif np.sign(f(b)) == np.sign(f(m)):
          return bisection(f, a, m, tol)
```

```python
# 5a

def f(x):
  return (x**-1 - tan(x))
r1=bisection(f,0.1,pi/2,1e-5)
print("r1 =",r1)
print("f(r1)",f(r1))
```

```python
# 5b
def f(x):
  return(x**-1 - 2**x)
r2=bisection(f,0.1,1,1e-5)
print("r2 =",r2)
print("f(r2)",f(r2))
```
